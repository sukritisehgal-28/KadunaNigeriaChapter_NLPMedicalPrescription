## Import Libraries

```python
## Import Libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import time

pd.set_option('display.max_colwidth', None) #enables columns to be displayed entirely
```

## Retrieve all webpages

### List all sub-pages in alpha

```python
import string
alphabets = string.ascii_lowercase

# Create list of all pairs of alphabets
url_list = [f"https://www.drugs.com/alpha/{first + second}.html" for first in alphabets for second in alphabets]

# Append to list all alphanumeric pages such as a0-9, b0-9 ,...
url_list.extend([f"https://www.drugs.com/alpha/{letter}0-9.html" for letter in alphabets])

# There should be 702 pages
print(len(url_list))
```

```python
# Code Demostration only
test = "https://www.drugs.com/alpha/A0-9.html"
pattern = r"/alpha/([A-Za-z0-9\-]+)\.html"

# Use re.findall() to find all matches
matches = re.findall(pattern, test)

print(matches[0])
print(matches[0][0])
```

```python
drug_list = list()
```

```python
num_of_pages = len(url_list)
pattern = r"/alpha/([A-Za-z0-9\-]+)\.html"

for page_index in range(num_of_pages):
  # First Alphabet
  first_alphabet = re.findall(pattern, url_list[page_index])[0][0]

  r = requests.get(url_list[page_index])

  if r.status_code == 200:
      page = r.content

  else:
      print('Some problem occurred. Request Status Code: %d' % r.status_code)

  soup = BeautifulSoup(page, 'html.parser')
  # print('Title tag :%s:' % soup.title)
  # print('Title text:%s:' % soup.title.string)

  li_tag_list = soup.find_all('li')

  for tag in li_tag_list:
    if tag.find("a") and "alpha" not in tag.find("a").get("href") and tag.text[0] == first_alphabet.upper():
      drug_list.append((tag.text, "https://www.drugs.com" + tag.find("a").get("href")))

  # Pause after every 20 pages
  if (page_index + 1) % 20 == 0:
    print("Pausing for 10 seconds...")
    time.sleep(10)

drug_list
```

```python
df = pd.DataFrame(drug_list, columns=['drug', 'URL'])
df
```

```python
df.to_csv("drugs.csv")
```
